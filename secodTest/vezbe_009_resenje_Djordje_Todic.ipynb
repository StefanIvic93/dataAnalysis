```python
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(close)
pyplot.show()

#Mozemo reƒái da se velika zavisnost javlja po≈°to vrednosti autokorelacije te≈æe vrednosti 1.
#Grafik autokorelacije.
```

```python
from pandas import DataFrame
from pandas import concat
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
values=DataFrame(close_2018.values)
dataframe=concat([values.shift(1),values],axis=1)
dataframe.columns=['t','t+1']
print(dataframe.head(5))
X=dataframe.values
train,test = X[1:len(X)-7],X[len(X)-7:]
train_X,train_y=train[:,0],train[:,1]
test_X,test_y=test[:,0],test[:,1]
def model_persistence(x):                            #Primer persistance modela,train i test seta.
    return x
predictions= list()
for x in train_X:
    yhat=model_persistence(x)
    predictions.append(yhat)
rmse=np.sqrt(mean_squared_error(train_y,predictions))
print('Test RMSE: %.3f '%rmse)
pyplot.plot(train_y)
pyplot.plot(predictions,color='red')
pyplot.show()
#Ovim smo pokazali da kretanje bitcoin-a mo≈æemo pretpostaviti jedan dan unapred.
```

```python
vektor_greske= train_y - predictions           #Vrednosti gre≈°ke predikcije
```

```python
pyplot.hist(vektor_greske)
```

```python
autocorrelation_plot(vektor_greske)        #Autokorelacioni grafik za vrednost gre≈°ke.
pyplot.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from random import randrange
close_2017=[i+randrange(10)for i in range(1,100)]
result=seasonal_decompose(close_2017,model='additive',freq=int(len(close_2017)/7))
result.plot()
pyplot.show()
#Aditivna dekompozicija na nedeljnom nivou.
```

```python
result=seasonal_decompose(close_2017,model='multiplicative',freq=int(len(close_2017)/7))
result.plot()
pyplot.show()
#Multiplikativna dekompozicija na nedeljnom nivou.
```
